```python
import folium
import pandas as pd
```

```python
# import df
station_df = pd.read_csv('raw_data/station_data.csv')
```

```python
# plot stations on a map

# select only reqired columns
station_locations = station_df[['name', 'latitude', 'longitude']]

# set mid long and lat locations to centre map
mid_lat = (station_locations.latitude.max() - station_locations.latitude.min())/2 + station_locations.latitude.min()
mid_log = (station_locations.longitude.max() - station_locations.longitude.min())/2 + station_locations.longitude.min()

# create map
station_map = folium.Map(location=[mid_lat, 
                                   mid_log], 
                                   zoom_start=6, 
                                   control_scale=True)

# plot points on the map
for index, location_info in station_locations.iterrows():
    folium.Marker([location_info["latitude"], location_info["longitude"]], popup=location_info["name"]).add_to(station_map)
    
station_map
```
